# 🌍 `locals()` та `globals()` — простори імен у Python

Функції `locals()` та `globals()` — це вбудовані інструменти **інтроспекції**, які дозволяють заглянути "під капот" вашої програми.

Вони показують:
* які змінні зараз існують
* де саме вони існують
* у якому просторі імен вони зберігаються

Обидві функції повертають **словник (`dict`)**, де:
* 🔑 ключі — це імена змінних (рядки)
* 📦 значення — це об’єкти, на які ці імена посилаються

---

# 🧠 Що таке простір імен (namespace)?

**Namespace** — це словник, який зв'язує:
`ім’я → об’єкт у пам’яті`

У Python існує кілька рівнів:
1. Built-in namespace
2. Global namespace
3. Local namespace

---

# 🔎 1️⃣ `globals()` — глобальний простір імен

Повертає словник змінних, функцій і класів, визначених на рівні модуля (файлу).

Глобальні змінні:
* існують весь час роботи програми
* доступні з будь-якого місця модуля

---

На верхньому рівні скрипта:

````python
locals() == globals()
````

Бо ми ще не всередині функції.

Типовий фрагмент результату:
```python
{
 '__name__': '__main__',
 '__doc__': None,
 'a': 10,
 'b': 20,
 ...
}
```

---

# 🔎 2️⃣ `locals()` — локальний простір імен

Повертає словник змінних, які існують **у поточній функції**.

Локальний простір:
* створюється при виклику функції
* знищується після її завершення

---

### Приклад

````python
global_var = 100

def my_function(param):
    local_var = "Я всередині"

    print("--- locals() ---")
    print(locals())

    print("\n--- globals() ---")
    print(globals())

my_function(42)
````

### Результат:
`locals()` покаже:
```python
{'param': 42, 'local_var': 'Я всередині'}
```

`globals()` покаже:
* всі глобальні змінні
* системні змінні
* саму функцію `my_function`
* `global_var`

---

# ⚠️ Важлива пастка
## ❗ `globals()` можна змінювати
## ❗ `locals()` — ні

### 🔹 Можна створити змінну через `globals()`

````python
globals()['new_var'] = 999
print(new_var)
````

Це реально створить глобальну змінну.

---

### 🔹 Але `locals()` працює лише для читання

````python
def test():
    locals()['x'] = 5
    print(locals())

test()
````

Змінна `x` НЕ стане реальною локальною змінною.
Бо:
> `locals()` повертає копію словника локального простору.

Щоб змінити локальну змінну — потрібно звичайне присвоювання:
```python
x = 5
```

# 🛠 Практичне використання

## 1️⃣ Швидкий дебагінг
Якщо функція працює неправильно:

````python
def complex_calculation(x, y, z):
    step1 = x * y
    step2 = step1 / z
    print(locals())
    return step2
````

Це дозволяє побачити весь стан функції в один момент.

---

## 2️⃣ Контроль `eval()` та `exec()`
`eval()` і `exec()` можуть приймати власні словники `globals` та `locals`.

````python
math_formula = "x * y + 5"

my_context = {'x': 10, 'y': 2}

result = eval(math_formula, {}, my_context)
print("Результат:", result)
````

Тут ми повністю контролюємо, до яких змінних має доступ код.

---

## 3️⃣ Форматування рядків (старий трюк)

````python
def create_greeting(name, age, city):
    return "Привіт! Мене звати %(name)s, мені %(age)s років, я з міста %(city)s." % locals()

print(create_greeting("Ivan", 30, "Kyiv"))
````

`locals()` передає всі змінні функції у форматування.
> ⚠️ У сучасному Python частіше використовують f-рядки.

# 🧬 Глибоке розуміння

Python реалізує правило:
## LEGB
1. **L**ocal
2. **E**nclosing
3. **G**lobal
4. **B**uilt-in

Коли Python шукає змінну — він проходить ці рівні по черзі.

---

# 🏗 Архітектурна модель
```text
Глобальний namespace
│
├── функція()
│     ├── локальний namespace
│     └── параметри
│
└── інші глобальні змінні
```

---

# 🎯 Коли реально використовувати?

| Ситуація | Чи варто використовувати |
| --- | --- |
| Дебагінг | ✅ Так |
| Метапрограмування | ✅ Так |
| Звичайна логіка | ❌ Ні |
| Створення змінних динамічно | ⚠️ Обережно |

---

# 🚫 Чого не варто робити
- ❌ Не будувати логіку програми через зміну `globals()`
- ❌ Не намагатися модифікувати `locals()`
- ❌ Не використовувати для "хитрих трюків" у звичайному коді

---

# 💡 Фінальний висновок
`globals()` та `locals()` — це:
> 🔦 Інструменти дослідження середовища виконання

Вони не для звичайної бізнес-логіки.
Вони для: дебагу, розуміння Python, метапрограмування, навчання.

🧠 LEGB — Як Python шукає змінні

Коли Python бачить ім’я змінної, він шукає її у строгому порядку:
🔤 **L → E → G → B**

| Рівень | Що означає | Де знаходиться |
| --- | --- | --- |
| **L** (Local) | Всередині поточної функції | 
| **E** (Enclosing)| У зовнішній функції (якщо вкладені) | 
| **G** (Global) | На рівні модуля (файлу) | 
| **B** (Built-in) | Вбудовані імена Python (len, print, str) |



🏗 **Візуальна схема**
```text
┌──────────────────────────────┐
│        Built-in (B)          │
│  print, len, str, int, ...   │
└──────────────▲───────────────┘
               │
┌──────────────┴───────────────┐
│        Global (G)            │
│  змінні модуля, функції      │
└──────────────▲───────────────┘
               │
┌──────────────┴───────────────┐
│     Enclosing (E)            │
│  зовнішня функція            │
└──────────────▲───────────────┘
               │
┌──────────────┴───────────────┐
│        Local (L)             │
│  поточна функція             │
└──────────────────────────────┘
```
Python починає з Local і рухається вгору.

---

🔎 1️⃣ **L — Local**

````python
x = "global"
def test():
    x = "local"
    print("Inside function:", x)

test()
print("Outside function:", x)
````

👉 Python спочатку шукає в Local → знаходить "local".

---

🔎 2️⃣ **E — Enclosing**

````python
def outer():
    x = "enclosing"
    def inner():
        print("Inner sees:", x)

    inner()
outer()
````

`inner()` не має власного `x`, тому Python піднімається до Enclosing.

---

🔎 3️⃣ **G — Global**

````python
x = "global variable"
def test():
    print("Inside function:", x)
test()
````

`x` не знайдено в Local → немає Enclosing → береться з Global.

---

🔎 4️⃣ **B — Built-in**

````python
print(len([1, 2, 3]))
````

`len` не визначений у Local, Enclosing чи Global, тому Python бере його з Built-in.

---

⚠️ **Пастка №1 — Перекриття Built-in**

````python
len = "oops"
print(len)

# print(len([1,2,3])) # ← тепер зламається
````

Ти перекрив built-in функцію `len`.

---

⚠️ **Пастка №2 — UnboundLocalError**

````python
x = 10
def test():
    print(x) # ❌ Це викличе помилку
    x = 5

test()
````

Python бачить, що всередині функції є присвоювання `x = 5`. Тому він вважає `x` локальною змінною, і не дозволяє звернутись до глобальної.

---

🛠 **Рішення — `global`**

````python
x = 10
def test():
    global x
    x = 5

test()
print(x)
````

🛠 **Рішення для Enclosing — `nonlocal`**

````python
def outer():
    x = 10
    def inner():
        nonlocal x
        x = 20

    inner()
    print(x)
outer()
````

🧬 **Глибока ментальна модель**
Коли Python бачить ім’я:
1. Чи є воно в Local?
2. Якщо ні — в Enclosing?
3. Якщо ні — в Global?
4. Якщо ні — в Built-in?
5. Якщо ні — `NameError`

🧠 **Ключова ідея**
LEGB — це не про "де змінна створена". Це про:
🔎 **Де Python її шукає.**

🚀 **Memory + LEGB разом**
`ім’я → namespace → об’єкт у пам’яті`

LEGB визначає:
* в якому namespace шукати
* яке ім’я пов’язане з яким об’єктом

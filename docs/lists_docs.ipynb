# Глибоке занурення: Списки (Lists), Методи та List Comprehensions

## 1. Що таке список (List) концептуально?



Уявіть список як безрозмірний контейнер або потяг, до якого можна під час руху чіпляти нові вагони, відчіпляти старі, міняти їх місцями або заглядати всередину будь-якого з них.

Списки в Python забезпечують підтримку впорядкованих колекцій інших об'єктів. Вони можуть містити елементи абсолютно різних типів (числа, рядки, інші списки тощо), і за необхідності автоматично збільшуватися або зменшуватися в розмірах.

**Головна роль списку — накопичення даних.**
Коли програма читає рядки з текстового файлу, збирає статистику з датчиків, отримує відповіді від користувачів або парсить веб-сайт, вона зазвичай складає всі ці шматочки інформації саме в список, щоб потім обробити їх усі разом.

## 2. Змінність (Mutability): Головна суперсила списків



Рядки (`str`) та кортежі (`tuple`) є незмінними. Якщо ви хочете змінити рядок, Python створює в пам'яті абсолютно новий рядок.
Списки (`list`) — це **змінний (mutable)** тип даних. Це означає, що ви можете змінювати вміст списку (додавати, видаляти, переписувати елементи) безпосередньо в тій самій ділянці пам'яті, де він був створений.

### Чому це важливо розуміти? (Пастка з посиланнями)
У Python змінні не зберігають самі списки безпосередньо, вони зберігають лише **посилання** на них. Оскільки список змінюється "на місці", це призводить до такого ефекту:

```python
a = [1, 2, 3]  # Створюємо список
b = a          # Змінна 'b' починає вказувати на ТОЙ САМИЙ список, що й 'a'
a.append(4)    # Ми змінюємо список через 'a'

print(b)       # Результат: [1, 2, 3, 4] - 'b' теж змінилося!
```

Саме тому списки ідеальні для накопичення: ви можете передати список у функцію, функція додасть туди нові дані, і ці зміни миттєво стануть доступними усій програмі.

## 4. Основні методи списків (для управління та накопичення)

### 🟢 Розширення та вставка (Накопичення)

**`append(x)`**
* **Опис:** Додає переданий елемент `x` у кінець існуючого списку. Це "мутуючий" метод — він змінює список in-place і повертає `None`.
* **Логіка:** Динамічне формування даних у циклах `for` або `while`.

**`extend(iterable)`**
* **Опис:** На відміну від `append`, `extend` бере іншу послідовність (список, рядок) і додає кожен її елемент по черзі в кінець поточного списку.
* **Логіка:** Використовується для злиття масивів даних.

**`insert(index, item)`**
* **Опис:** Вставляє елемент `item` на вказану позицію `index`. Усі наступні елементи зсуваються вправо.
* **Логіка:** Коли дані треба додати не в кінець, а в конкретне місце (наприклад, додати термінове завдання на початок черги).

```python
# append
cars = ['Toyota', 'BMW']
cars.append('Audi')
print("Append:", cars)

# extend
cart = ['яблуко', 'банан']
new_items = ['ківі', 'манго']
cart.extend(new_items)
print("Extend:", cart) # Результат: ['яблуко', 'банан', 'ківі', 'манго']

# insert
queue = ['Олег', 'Анна']
queue.insert(0, 'Бос') 
print("Insert:", queue) # Результат: ['Бос', 'Олег', 'Анна']
```

### 🔴 Видалення елементів (Очищення даних)

**`remove(item)`**
* **Опис:** Шукає перше входження значення `item` і видаляє його. Якщо такого елемента немає, видає помилку `ValueError`.
* **Логіка:** Видалення конкретного об'єкта, коли ви знаєте *що* треба видалити, але не знаєте *де* воно лежить.

**`pop([index])`**
* **Опис:** Видаляє елемент за вказаним індексом і **повертає** його. Якщо індекс не вказано, видаляє і повертає останній елемент.
* **Логіка:** Основа для створення "стеків" або "черг". Ви одночасно забираєте дані зі списку і починаєте з ними працювати.

**`clear()`**
* **Опис:** Повністю очищає список, видаляючи з нього всі елементи. Список стає порожнім `[]`.
* **Логіка:** Коли ви накопичили дані, обробили/зберегли їх, і вам потрібен чистий контейнер для наступної порції.

```python
# remove
colors = ['red', 'blue', 'green', 'blue']
colors.remove('blue')
print("Remove:", colors) # Результат: ['red', 'green', 'blue']

# pop
tasks = ['Task1', 'Task2']
current_task = tasks.pop()
print("Pop повернув:", current_task) # 'Task2'
print("Залишилось у tasks:", tasks)   # ['Task1']

# clear
tasks.clear()
print("Clear:", tasks) # []
```

### 🔎 Пошук та аналіз

**`index(item, [start, end])`**
* **Опис:** Повертає індекс (позицію) першого входження елемента `item`.
* **Логіка:** Пошук місцезнаходження даних для подальших маніпуляцій.

**`count(item)`**
* **Опис:** Підраховує, скільки разів елемент `item` зустрічається у списку.
* **Логіка:** Швидка статистика (наприклад, скільки разів користувач натиснув певну кнопку).

```python
data = ['a', 'b', 'c', 'b', 'b']
print("Index of 'c':", data.index('c')) # Результат: 2
print("Count of 'b':", data.count('b')) # Результат: 3
```

### 🔄 Впорядкування (Мутація "на місці")

**`sort(key=None, reverse=False)`**
* **Опис:** Сортує елементи списку. Дуже важливо: на відміну від функції `sorted()`, яка створює новий список, метод `.sort()` змінює сам поточний список (in-place) і повертає `None`.
* **Логіка:** Організація накопичених даних перед виведенням.

**`reverse()`**
* **Опис:** Розвертає порядок елементів у списку задом наперед. Змінює список на місці.

```python
# sort
prices = [150, 20, 99, 5]
prices.sort(reverse=True) # Сортування за спаданням
print("Sort:", prices) # Результат: [150, 99, 20, 5]

# reverse
items = [1, 2, 3]
items.reverse()
print("Reverse:", items) # Результат: [3, 2, 1]
```

## 5. Спискові включення (List Comprehensions)



**Опис:**
Спискове включення (List Comprehension) — це елегантний і швидкий спосіб створення нового списку з ітерованого об'єкта шляхом застосування виразу до кожного елемента та/або фільтрації.
Це "синтаксичний цукор", який дозволяє замінити громіздкі цикли `for` з `append()` одним читабельним рядком коду.

**Загальний синтаксис:**
`[вираз for змінна in ітерований_об'єкт if умова]`

**Логіка використання:**
* **Трансформація даних:** Зміна кожного елемента списку (аналог `map()`).
* **Фільтрація даних:** Вибір елементів, що відповідають умові (аналог `filter()`).
* **Оптимізація:** Працюють швидше, ніж звичайний цикл `for` з `append()`, оскільки оптимізовані на рівні С.

### А. Трансформація (заміна `for` + `append`)

```python
numbers = [1, 2, 3]
doubled = [n * 2 for n in numbers]
print("Doubled:", doubled) # Результат: [2, 4, 6]
```

### Б. Фільтрація (використання `if`)

```python
nums = [1, 2, 3, 4, 5, 6]
evens = [n for n in nums if n % 2 == 0]
print("Evens:", evens) # Результат: [2, 4, 6]
```

### В. Робота з рядками

```python
words = ["Apple", "Banana", "Cherry"]
first_letters = [word[0] for word in words] 
print("First letters:", first_letters) # Результат: ['A', 'B', 'C']
```

### Г. Декартовий добуток (вкладені цикли `for`)

```python
colors = ['red', 'blue']
items = ['car', 'pen']
combinations = [f"{c} {i}" for c in colors for i in items]
print("Combinations:", combinations)
# Результат: ['red car', 'red pen', 'blue car', 'blue pen']
```

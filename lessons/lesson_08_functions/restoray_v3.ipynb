```python
from typing import NamedTuple

from generator.restoran_v4 import smoker


class Order(NamedTuple):
    total_bill: float
    tip: float
    sex: str
    smoker: str
    day: str
    time: str
    size: int
```

```python
import seaborn as sns



tips_dfm = sns.load_dataset("tips")
print(tips_dfm.head())
```

```python
def  load_df(df):
    orders = [
    Order(
        total_bill = row["total_bill"],
        tip = row["tip"],
        sex= row["sex"],
        smoker= row["smoker"],
        day= row["day"],
        time= row["time"],
        size = row["size"]
    )
for _, row in df.iterrows()]
    return orders

```

```python
orders = load_df(tips_dfm)
```

```python
smoker_geust= []
for o in orders:
    if o.smoker == "Yes":\
        smoker_geust.append(o)
print(smoker_geust)
```

```python
def avg_bill(orders):
    if not orders:
        return 0.0
    total_bill = sum(o.total_bill for o in orders)
    print(total_bill)
    avg_bill = total_bill / len(orders)
    print(avg_bill)
    return avg_bill

```

```python
sat_orders = [x for x in orders if x.day == "Sat"]
print(avg_bill(sat_orders))
```

```python
sat_orders = [x for x in orders if x.time == "Dinner"]
print(avg_bill(sat_orders))
```

```python
sat_orders = [x for x in orders if x.size >= 3]
print(avg_bill(sat_orders))
```

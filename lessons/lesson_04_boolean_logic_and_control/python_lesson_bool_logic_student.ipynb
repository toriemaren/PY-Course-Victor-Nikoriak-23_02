````python
import ipywidgets as widgets
from IPython.display import display, Markdown

_rev2_btn = widgets.Button(description="🐍 Результат та пояснення", button_style="info")
_rev2_out = widgets.Output()

def _on_rev2(b):
    if not require_student():
        with _rev2_out:
            print("⛔ Спочатку представтесь")
        return
    if not prediction_logic.strip():
        with _rev2_out:
            print("⚠️ Спочатку зробіть прогноз у клітинці 🔮")
        return
    b.disabled = True
    with _rev2_out:
        print(f"{STUDENT_NAME}, твій прогноз: {prediction_logic}")

        x, y = True, False
        result = not x and y or x

        print("\n🐍 Покроково:")
        print("x =", x)
        print("y =", y)
        print("not x        =", not x)
        print("not x and y  =", not x and y)
        print("... or x     =", not x and y or x)
        print("\nresult =", result)

        correct_answer = "True"
        if prediction_logic.strip() == correct_answer:
            print("\n✅ Правильний прогноз!")
        else:
            print("\n❌ Неправильно. Очікувалось:", correct_answer)

        display(Markdown("""
---

## ✅ Що сталося?

### Пріоритет: `not` > `and` > `or`

Вираз `not x and y or x` читається як `((not x) and y) or x`

| Крок | Вираз              | Значення |
|------|--------------------|----------|
| 1    | `not True`         | `False`  |
| 2    | `False and False`  | `False`  |
| 3    | `False or True`    | `True`   |

---

### Short-circuiting

```python
True or f()    # f() НЕ викликається
False and f()  # f() НЕ викликається
```

## 🔑 Правило

> Якщо сумніваєтесь — **ставте дужки**. Код стане зрозумілішим.
"""))

        try:
            submit_task(STUDENT_NAME, "lesson_04_exam", "Task_logic_operators")
        except Exception:
            print("\nℹ️ submit_task() не підключений")

_rev2_btn.on_click(_on_rev2)
display(widgets.VBox([_rev2_btn, _rev2_out]))
````

# ✅ Task 4 — if / elif / else

### Що виведе Python?

```python
score = 85
if score >= 60:
    print("Ви склали іспит!")
```

> ✏️ Запишіть **точну фразу** яку виведе Python, перш ніж запускати наступну клітинку.

````python
import ipywidgets as widgets
from IPython.display import display, clear_output

prediction_if = ""

_pb4_w = widgets.Text(
    placeholder="Ваш прогноз",
    description="Прогноз:",
    layout=widgets.Layout(width="320px"))
_pb4_btn = widgets.Button(description="Зберегти прогноз", button_style="primary")
_pb4_out = widgets.Output()

def _on_pb4(b):
    global prediction_if
    if not require_student():
        with _pb4_out:
            clear_output()
            print("⛔ Спочатку представтесь")
        return
    val = _pb4_w.value.strip()
    if not val:
        with _pb4_out:
            clear_output()
            print("⚠️ Введіть прогноз перед виконанням наступної клітинки")
        return
    prediction_if = val
    _pb4_btn.disabled = True
    with _pb4_out:
        clear_output()
        print("✅ Прогноз збережено:", prediction_if)

_pb4_btn.on_click(_on_pb4)
display(widgets.VBox([_pb4_w, _pb4_btn, _pb4_out]))
````

````python
import ipywidgets as widgets
from IPython.display import display, Markdown
import io, sys

_rev4_btn = widgets.Button(description="🐍 Результат та пояснення", button_style="info")
_rev4_out = widgets.Output()

def _on_rev4(b):
    if not require_student():
        with _rev4_out:
            print("⛔ Спочатку представтесь")
        return
    if not prediction_if.strip():
        with _rev4_out:
            print("⚠️ Спочатку зробіть прогноз у клітинці 🔮")
        return
    b.disabled = True
    with _rev4_out:
        print(f"{STUDENT_NAME}, твій прогноз: {prediction_if}")

        # Capture stdout without losing the Output widget stream
        _prev_stdout = sys.stdout
        buf = io.StringIO()
        sys.stdout = buf

        score = 85
        if score >= 60:
            print("Ви склали іспит!")
        elif score >= 80:
            print("Ви отримали оцінку B!")
        else:
            print("Іспит не складено.")

        sys.stdout = _prev_stdout
        actual_output = buf.getvalue().strip()

        print("\n🐍 Python повернув:", actual_output)

        correct_answer = "Ви склали іспит!"
        if prediction_if.strip() == correct_answer:
            print("\n✅ Правильний прогноз!")
        else:
            print("\n❌ Неправильно. Очікувалось:", correct_answer)

        display(Markdown("""
---

## ✅ Що сталося?

### Перевірка зверху вниз

```python
score = 85
if score >= 60:    # 85 >= 60 → True ✅ Виконується!
    print("Ви склали іспит!")
elif score >= 80:  # ← ніколи не перевіряється
    print("Ви отримали оцінку B!")
```

Python знаходить **першу** умову `True`, виконує блок і **ігнорує** решту.

```python
# ✅ Правильний порядок — від строгішого до загального
if score >= 90:   print("Відмінно")
elif score >= 80: print("Добре")
elif score >= 60: print("Склав")
else:             print("Не склав")
```

## 🔑 Головне правило

> У `if-elif-else` виконується **лише перший** блок з True-умовою.  
> Розміщуйте умови від **конкретних** до **загальних**.
"""))

        try:
            submit_task(STUDENT_NAME, "lesson_04_exam", "Task_if_elif")
        except Exception:
            print("\nℹ️ submit_task() не підключений")

_rev4_btn.on_click(_on_rev4)
display(widgets.VBox([_rev4_btn, _rev4_out]))
````

# 5. Цикл while та нескінченні цикли

## ⚠️ Знайдіть помилку (не запускайте!)

```python
x = 1
while x <= 5:
    print(x)
```

Що не так? `x` ніколи не змінюється → умова `1 <= 5` завжди `True` → **нескінченний цикл**.

**Виправлений варіант:**

```python
x = 1
while x <= 5:
    print(x)
    x += 1   # ← обов'язково!
```

````python
# Правильний while: рахуємо до 5
x = 1
while x <= 5:
    print(x)
    x += 1

print("Готово!")
````

# ✅ Task 5 — break та continue

### Що виведе Python?

```python
for num in range(1, 6):
    if num == 3:
        continue
    if num == 5:
        break
    print(num)
```

> ✏️ Введіть числа через пробіл (наприклад: `1 2 4`).

````python
import ipywidgets as widgets
from IPython.display import display, clear_output

prediction_loop = ""

_pb5_w = widgets.Text(
    placeholder="Ваш прогноз",
    description="Прогноз:",
    layout=widgets.Layout(width="320px"))
_pb5_btn = widgets.Button(description="Зберегти прогноз", button_style="primary")
_pb5_out = widgets.Output()

def _on_pb5(b):
    global prediction_loop
    if not require_student():
        with _pb5_out:
            clear_output()
            print("⛔ Спочатку представтесь")
        return
    val = _pb5_w.value.strip()
    if not val:
        with _pb5_out:
            clear_output()
            print("⚠️ Введіть прогноз перед виконанням наступної клітинки")
        return
    prediction_loop = val
    _pb5_btn.disabled = True
    with _pb5_out:
        clear_output()
        print("✅ Прогноз збережено:", prediction_loop)

_pb5_btn.on_click(_on_pb5)
display(widgets.VBox([_pb5_w, _pb5_btn, _pb5_out]))
````

````python
import ipywidgets as widgets
from IPython.display import display, Markdown

_rev5_btn = widgets.Button(description="🐍 Результат та пояснення", button_style="info")
_rev5_out = widgets.Output()

def _on_rev5(b):
    if not require_student():
        with _rev5_out:
            print("⛔ Спочатку представтесь")
        return
    if not prediction_loop.strip():
        with _rev5_out:
            print("⚠️ Спочатку зробіть прогноз у клітинці 🔮")
        return
    b.disabled = True
    with _rev5_out:
        print(f"{STUDENT_NAME}, твій прогноз: {prediction_loop}")

        print("\n🐍 Python виводить (покроково):")
        for num in range(1, 6):
            if num == 3:
                print(f"  num={num} → continue (пропускаємо)")
                continue
            if num == 5:
                print(f"  num={num} → break (зупиняємось)")
                break
            print(f"  num={num} → print({num})")

        print("\n📌 Фактичний вивід:")
        for num in range(1, 6):
            if num == 3:
                continue
            if num == 5:
                break
            print(num)

        correct_answer = "1 2 4"
        if prediction_loop.strip() == correct_answer:
            print("\n✅ Правильний прогноз!")
        else:
            print("\n❌ Неправильно. Очікувалось:", correct_answer)

        display(Markdown("""
---

## ✅ Що сталося?

| Ітерація | num | Дія          | Виводить? |
|----------|-----|--------------|----------|
| 1        | 1   | звичайна     | ✅ `1`   |
| 2        | 2   | звичайна     | ✅ `2`   |
| 3        | 3   | `continue`   | ❌ пропуск |
| 4        | 4   | звичайна     | ✅ `4`   |
| 5        | 5   | `break`      | ❌ вихід |

**Результат:** `1 2 4`

```python
# break — повністю зупиняє цикл
# continue — пропускає лише цю ітерацію
```

## 🔑 Головне правило

> `break` = вийти з циклу назавжди.  
> `continue` = перейти до наступної ітерації.
"""))

        try:
            submit_task(STUDENT_NAME, "lesson_04_exam", "Task_break_continue")
        except Exception:
            print("\nℹ️ submit_task() не підключений")

_rev5_btn.on_click(_on_rev5)
display(widgets.VBox([_rev5_btn, _rev5_out]))
````

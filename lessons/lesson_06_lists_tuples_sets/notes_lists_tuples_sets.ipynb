## 🧱 Частина 1: Фундаментальні відмінності

Перед тим як писати код, потрібно зрозуміти три головні контейнери Python та їхню ключову характеристику — **змінюваність (mutability)**.

```
┌─────────────────────────────────────────────────────────────────────┐
│             ПОРІВНЯННЯ КОНТЕЙНЕРІВ                                  │
│                                                                     │
│  LIST  []          TUPLE  ()          SET  {}                       │
│  ─────────         ──────────         ──────────                    │
│  Впорядкований     Впорядкований      Невпорядкований               │
│  Змінний           Незмінний          Змінний                       │
│  Дублікати ✅      Дублікати ✅       Дублікати ❌ (унікальні)       │
│  Індекс    ✅      Індекс    ✅       Індекс    ❌                   │
│                                                                     │
│  [1, 2, 2, 3]      (1, 2, 2, 3)       {1, 2, 3}                     │
└─────────────────────────────────────────────────────────────────────┘
```

**Коли що використовувати?**

| Структура | Використовуйте коли... | Приклад |
|-----------|------------------------|---------|
| **Список `[]`** | Дані динамічні, порядок важливий | Кошик покупок, список ворогів у грі |
| **Кортеж `()`** | Дані фіксовані, захист від змін | Координати `(x, y)`, налаштування |
| **Множина `{}`** | Потрібна унікальність, швидка перевірка | Унікальні відвідувачі, теги статті |

````python
# Приклад 1.1: Наочне порівняння трьох контейнерів

raw_data = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3]  # дані з дублікатами

my_list  = list(raw_data)       # список — зберігає всі елементи та порядок
my_tuple = tuple(raw_data)      # кортеж — те саме, але незмінний
my_set   = set(raw_data)        # множина — тільки унікальні

print(f"Вхідні дані : {raw_data}")
print(f"list  {list.__name__:<5}: {my_list}")
print(f"tuple {tuple.__name__:<5}: {my_tuple}")
print(f"set   {set.__name__:<5}: {my_set}")
print()

print(f"Довжини: list={len(my_list)}, tuple={len(my_tuple)}, set={len(my_set)}")
print(f"Дублікати прибрані: {len(my_list) - len(my_set)} шт.")
print()

# Перевірка змінюваності
my_list[0] = 999        # ✅ ок
print(f"list після зміни: {my_list}")

try:
    my_tuple[0] = 999   # ❌ кортеж незмінний
except TypeError as e:
    print(f"tuple: TypeError — {e}")

try:
    _ = my_set[0]       # ❌ множина не підтримує індексацію
except TypeError as e:
    print(f"set:   TypeError — {e}")
````

````python
# Приклад 1.2: Порівняння витрат пам'яті
import sys

data = list(range(1000))

lst = list(data)
tpl = tuple(data)
st  = set(data)

print("=== Витрати пам'яті (1000 елементів) ===")
print(f"  list  : {sys.getsizeof(lst):>8} байт")
print(f"  tuple : {sys.getsizeof(tpl):>8} байт")
print(f"  set   : {sys.getsizeof(st):>8} байт")
print()
print("Висновок: tuple займає найменше пам'яті.")
print("set потребує більше пам'яті через хеш-таблицю, але дає O(1) пошук.")
````

## 📋 Частина 2: Списки — детальний розбір

**Список (list)** — впорядкована, **змінювана** колекція довільних об'єктів.

### Головні характеристики

- **Змінність (Mutability):** список може збільшуватись або зменшуватись прямо в пам'яті
- **Різнорідність:** один список може містити числа, рядки, інші списки
- **Вкладеність:** списки можуть містити інші списки (матриці)
- **Впорядкованість:** елементи мають чітке позиційне впорядкування зліва направо

### Індексація та зрізи

```
 Список:    ['яблуко', 'банан', 'апельсин', 'ківі', 'манго']
 Індекс +:      0        1          2          3       4
 Індекс -:     -5       -4         -3         -2      -1
```

```python
fruits[1:4]   # ['банан', 'апельсин', 'ківі']  — від 1 до 3 включно
fruits[:3]    # ['яблуко', 'банан', 'апельсин'] — від початку до 2
fruits[2:]    # ['апельсин', 'ківі', 'манго']   — від 2 до кінця
fruits[::2]   # ['яблуко', 'апельсин', 'манго'] — кожен другий
fruits[::-1]  # ['манго', 'ківі', 'апельсин', 'банан', 'яблуко'] — реверс
```

````python
# Приклад 2.1: Індексація та зрізи
fruits = ['яблуко', 'банан', 'апельсин', 'ківі', 'манго']

print("=== Індексація ===")
print(f"fruits[0]   = {fruits[0]!r:<15}  (перший)")
print(f"fruits[2]   = {fruits[2]!r:<15}  (третій)")
print(f"fruits[-1]  = {fruits[-1]!r:<15}  (останній)")
print(f"fruits[-2]  = {fruits[-2]!r:<15}  (передостанній)")
print()

print("=== Зрізи ===")
print(f"fruits[1:4]   = {fruits[1:4]}")
print(f"fruits[:3]    = {fruits[:3]}")
print(f"fruits[2:]    = {fruits[2:]}")
print(f"fruits[::2]   = {fruits[::2]}")
print(f"fruits[::-1]  = {fruits[::-1]}")
print()

print("=== Зміна через зріз ===")
copy = fruits[:]            # копія списку
copy[1:3] = ['БАНАН', 'АП'] # заміна зрізу
print(f"copy після copy[1:3] = [...]: {copy}")
````

````python
# Приклад 2.2: Основні методи списку
fruits = ['яблуко', 'банан', 'апельсин']

print("=== Додавання ===")
fruits.append('ківі')           # в кінець
print(f"append('ківі')     : {fruits}")

fruits.insert(1, 'манго')       # на позицію 1
print(f"insert(1,'манго')  : {fruits}")

fruits.extend(['диня', 'слива']) # злиття зі списком
print(f"extend([...])      : {fruits}")
print()

print("=== Видалення ===")
fruits.remove('банан')           # за значенням (перше входження)
print(f"remove('банан')    : {fruits}")

popped = fruits.pop()            # з кінця — і повертає елемент
print(f"pop()={popped!r:<10}: {fruits}")

popped2 = fruits.pop(0)          # з індексу 0
print(f"pop(0)={popped2!r:<8}: {fruits}")
print()

print("=== Пошук та підрахунок ===")
sample = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3]
print(f"sample.count(1)    = {sample.count(1)}  (скільки разів 1 зустрічається)")
print(f"sample.index(5)    = {sample.index(5)}  (індекс першого 5)")
print(f"len(sample)        = {len(sample)}")
print(f"min(sample)        = {min(sample)}")
print(f"max(sample)        = {max(sample)}")
print(f"sum(sample)        = {sum(sample)}")
print()

print("=== Сортування ===")
nums = [3, 1, 4, 1, 5, 9, 2, 6]
sorted_copy = sorted(nums)       # повертає НОВИЙ список
print(f"sorted(nums)       = {sorted_copy}  (оригінал не змінений)")
nums.sort()                      # сортує НА МІСЦІ
print(f"nums після .sort() = {nums}")
nums.sort(reverse=True)
print(f"nums sort(desc)    = {nums}")
nums.reverse()                   # реверс на місці
print(f"nums після reverse = {nums}")
````

````python
# Приклад 2.3: Спискові включення (List Comprehensions)
# Синтаксис: [вираз for елемент in послідовність if умова]

vec = [2, 4, 6, 8, 10]

print("=== Базові включення ===")
tripled = [3 * x for x in vec]
print(f"vec              = {vec}")
print(f"[3*x for x in v] = {tripled}")
print()

squares = [[x, x**2] for x in vec]
print(f"[[x, x**2] ...] = {squares}")
print()

print("=== Включення з умовою (фільтрація) ===")
evens = [x for x in range(1, 21) if x % 2 == 0]
print(f"Парні від 1 до 20 : {evens}")

long_words = [w for w in ['яблуко','ів','апельсин','ківі','банан'] if len(w) > 4]
print(f"Слова > 4 символи : {long_words}")
print()

print("=== Вкладені включення (матриця) ===")
matrix_3x3 = [[i * j for j in range(1, 4)] for i in range(1, 4)]
print("Таблиця множення 3×3:")
for row in matrix_3x3:
    print(f"  {row}")
print()

print("=== Порівняння: цикл vs включення ===")
# Еквівалентний результат різними способами
nums = range(1, 11)

# Варіант 1: звичайний цикл
result_loop = []
for n in nums:
    if n % 2 != 0:
        result_loop.append(n ** 2)

# Варіант 2: включення (одним рядком)
result_comp = [n**2 for n in nums if n % 2 != 0]

print(f"Квадрати непарних (цикл)      : {result_loop}")
print(f"Квадрати непарних (comprehension): {result_comp}")
print(f"Результати однакові: {result_loop == result_comp}")
````

````python
# Приклад 2.4: Вкладені списки (матриці)
print("=== Вкладені списки ===")

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print("Матриця:")
for row in matrix:
    print(" ", row)
print()

print(f"matrix[0]     = {matrix[0]}     (перший рядок)")
print(f"matrix[1][2]  = {matrix[1][2]}       (рядок 1, стовпець 2)")
print(f"matrix[-1]    = {matrix[-1]}  (останній рядок)")
print()

# Транспонування матриці через включення
transposed = [[row[i] for row in matrix] for i in range(3)]
print("Транспонована матриця:")
for row in transposed:
    print(" ", row)
````

## 📌 Частина 3: Кортежі — незмінні послідовності

**Кортеж (tuple)** — впорядкована, **незмінна** послідовність. Після створення не можна змінити жоден елемент.

### Чому кортеж, а не список?

| Причина | Пояснення |
|---------|----------|
| **Продуктивність** | Менше пам'яті, швидше створення |
| **Безпека** | Захист даних від випадкових змін |
| **Ключ словника** | `dict[(x, y)] = value` — tuple можна хешувати, list — ні |
| **Розпакування** | `x, y, z = point` — зручний синтаксис |

### Незмінність є «відносною»

```python
t = (1, 2, [10, 20])   # кортеж містить список
t[2].append(30)         # ✅ список всередині кортежу — можна змінити!
# t[2] = [99]           # ❌ замінити посилання — не можна
```

### `NamedTuple` — іменовані кортежі як замінники словників

```
Звичайний кортеж:     point[0], point[1]
                        ↑ що це? незрозуміло без контексту

NamedTuple:           point.x, point.y
                        ↑ самодокументований код!
```

````python
# Приклад 3.1: Базові операції з кортежами

# Створення
empty    = ()
single   = (42,)         # ОБОВ'ЯЗКОВА кома!
point    = (10, 20, 30)
colors   = 'Red', 'Green', 'Blue'   # без дужок — теж кортеж
from_seq = tuple([1, 2, 3])         # з послідовності

print(f"empty   = {empty}")
print(f"single  = {single}  ← (42) без коми = int 42, не кортеж!")
print(f"point   = {point}")
print(f"colors  = {colors}")
print(f"from_seq= {from_seq}")
print()

print("=== Методи та операції ===")
t = (3, 1, 4, 1, 5, 9, 2, 6, 5, 3)
print(f"t.count(1)   = {t.count(1)}    (кількість входжень)")
print(f"t.index(5)   = {t.index(5)}    (індекс першого 5)")
print(f"t[1:4]       = {t[1:4]}  (зріз — дає новий кортеж)")
print(f"(1,2)+(3,4)  = {(1,2)+(3,4)} (конкатенація)")
print(f"(1,)*3       = {(1,)*3}  (множення)")
print()

print("=== Незмінність є відносною ===")
mutable_inside = (1, 2, [10, 20])
mutable_inside[2].append(30)     # змінюємо список ВСЕРЕДИНІ кортежу
print(f"Після append: {mutable_inside}   ← список змінився!")
try:
    mutable_inside[0] = 99       # але сам кортеж — незмінний
except TypeError as e:
    print(f"TypeError: {e}")
````

````python
# Приклад 3.2: Пакування та розпакування

point = (10, 20, 30)

# Розпакування: кількість змінних = кількість елементів
x, y, z = point
print(f"x={x}, y={y}, z={z}")

# Розпакування з * (зірочкою) — для решти
first, *rest = (1, 2, 3, 4, 5)
print(f"first={first}, rest={rest}")

*head, last = (1, 2, 3, 4, 5)
print(f"head={head}, last={last}")

a, *middle, b = (1, 2, 3, 4, 5)
print(f"a={a}, middle={middle}, b={b}")
print()

# Обмін без тимчасової змінної
a, b = 10, 20
print(f"До обміну: a={a}, b={b}")
a, b = b, a
print(f"Після обміну: a={a}, b={b}")
print()

# Tuple як ключ словника
locations = {
    (50.45, 30.52): 'Київ',
    (48.46, 35.05): 'Дніпро',
    (49.99, 36.23): 'Харків',
}
coords = (50.45, 30.52)
print(f"Місто за координатами {coords}: {locations[coords]}")
````

````python
# Приклад 3.3: namedtuple (collections)
from collections import namedtuple

# Старий підхід: словник — ключі можна помилково написати
car_dict = {'color': 'red', 'mileage': 3812.4, 'year': 2021}
print("Словник:")
print(f"  car_dict['color']   = {car_dict['color']}")
print(f"  car_dict['mileage'] = {car_dict['mileage']}")
car_dict['colur'] = 'blue'   # опечатка — новий ключ, не помилка!
print(f"  car_dict після опечатки: {car_dict}")  # 'colur' тихо додався!
print()

# Новий підхід: namedtuple — імена полів фіксовані
Car = namedtuple('Car', 'color mileage year')
my_car = Car(color='red', mileage=3812.4, year=2021)

print("namedtuple:")
print(f"  my_car.color   = {my_car.color}")
print(f"  my_car.mileage = {my_car.mileage}")
print(f"  my_car[0]      = {my_car[0]}       ← доступ за індексом теж працює")
print(f"  my_car         = {my_car}")
print()

# Захист від опечаток
try:
    _ = my_car.colur   # опечатка → AttributeError
except AttributeError as e:
    print(f"  Захист від опечатки: AttributeError — {e}")

# Захист від змін
try:
    my_car.color = 'blue'
except AttributeError as e:
    print(f"  Захист від зміни: AttributeError — {e}")
print()

# Розпакування та конвертація в словник
color, mileage, year = my_car
print(f"  Розпакування: color={color}, mileage={mileage}, year={year}")
print(f"  _asdict():    {my_car._asdict()}")
````

````python
# Приклад 3.4: typing.NamedTuple — сучасний підхід з анотаціями типів
from typing import NamedTuple

class Coordinate(NamedTuple):
    latitude: float
    longitude: float
    label: str = 'Unknown'   # значення за замовчуванням

class Player(NamedTuple):
    name: str
    score: int
    level: int = 1

# Координати
kyiv = Coordinate(50.45, 30.52, 'Київ')
print(f"Координати: {kyiv}")
print(f"  latitude  = {kyiv.latitude}")
print(f"  longitude = {kyiv.longitude}")
print(f"  label     = {kyiv.label}")
print()

# Гравці
players = [
    Player('Alice', 1500, 10),
    Player('Bob', 2300, 15),
    Player('Carol', 900, 7),
]

print("Таблиця лідерів:")
for rank, p in enumerate(sorted(players, key=lambda p: p.score, reverse=True), 1):
    print(f"  {rank}. {p.name:<8} score={p.score:>5}  level={p.level}")
print()

print("=== Порівняння підходів ===")
print(f"  {'collections.namedtuple':<30}: Car = namedtuple('Car', 'color mileage')")
print(f"  {'typing.NamedTuple':<30}: class Car(NamedTuple): color: str")
print("  Обидва — незмінні, ефективні, підтримують розпакування.")
print("  typing.NamedTuple краще: анотації типів, значення за замовчуванням.")
````

## 🔵 Частина 4: Множини — унікальність та операції

**Множина (set)** — неупорядкована колекція **унікальних** об'єктів на основі хеш-таблиці.

### Механізм унікальності (під капотом)

- Елементи **мають бути хешованими** (незмінними): числа, рядки, кортежі — так; списки, словники — ні
- Перевірка `in` працює за **O(1)** — миттєво, незалежно від розміру
- **Порядок не гарантований** — не покладайтесь на нього

### Діаграми Венна — візуалізація операцій

```
  A = {1,2,3,4}    B = {3,4,5,6}

  UNION (об'єднання A | B)         INTERSECTION (перетин A & B)
  ╔═══════════╗                    ╔═══════════╗
  ║  ░░░░░░░░░║░░░░░░░║            ║           ║░░░║
  ║ 1 2 ░3 4░ ║ 5 6   ║            ║ 1 2  ░3 4░    5 6 ║
  ║  ░░░░░░░░░║░░░░░░░║            ║           ║░░░║
  A∪B = {1,2,3,4,5,6}              A∩B = {3,4}

  DIFFERENCE (різниця A - B)       SYMMETRIC DIFF (A ^ B)
  ╔═══════════╗                    ╔═══════════╗
  ║ ░░░       ║       ║            ║ ░░░       ║   ░░░░░║
  ║░1 2░ 3 4  ║ 5 6   ║            ║░1 2░ 3 4  ║  ░5 6░ ║
  ║ ░░░       ║       ║            ║ ░░░       ║   ░░░░░║
  A-B = {1,2}                      A^B = {1,2,5,6}
```

````python
# Приклад 4.3: Операції над трьома множинами
shop_A = {'молоко', 'хліб', 'масло', 'яйця', 'кава'}
shop_B = {'молоко', 'хліб', 'цукор', 'сіль', 'кава'}
shop_C = {'хліб', 'масло', 'цукор', 'олія', 'кава'}

print("=== Задача: Що є у всіх трьох магазинах? ===")
in_all = shop_A & shop_B & shop_C
print(f"shop_A & shop_B & shop_C = {in_all}")
print()

print("=== Що є хоча б в одному магазині? ===")
in_any = shop_A | shop_B | shop_C
print(f"shop_A | shop_B | shop_C = {in_any}")
print()

print("=== Що є в A, але відсутнє в B і C? ===")
only_in_A = shop_A - shop_B - shop_C
print(f"shop_A - shop_B - shop_C = {only_in_A}")
print()

# Через методи з кількома аргументами
print("=== Через методи (кілька аргументів) ===")
print(f"shop_A.intersection(shop_B, shop_C) = {shop_A.intersection(shop_B, shop_C)}")
print(f"shop_A.union(shop_B, shop_C) = {shop_A.union(shop_B, shop_C)}")
````

````python
# Приклад 4.4: add, remove, discard, pop — різниця
s = {10, 20, 30, 40, 50}
print(f"Початкова: {s}")
print()

# add — додає елемент (без помилки якщо вже є)
s.add(60)
s.add(30)    # вже є — нічого не відбувається
print(f"Після add(60), add(30): {s}")

# remove — видаляє, але кидає KeyError якщо немає
s.remove(10)
print(f"Після remove(10): {s}")
try:
    s.remove(999)
except KeyError as e:
    print(f"remove(999) → KeyError: {e}")

# discard — видаляє БЕЗПЕЧНО (немає → ігнорує)
s.discard(999)   # елемента немає — але помилки немає!
s.discard(20)
print(f"Після discard(999), discard(20): {s}")
print()

# pop — видаляє ДОВІЛЬНИЙ елемент і ПОВЕРТАЄ його
print("=== pop() — деструктивна ітерація ===")
work_set = {1, 2, 3, 4, 5}
print(f"Початок: {work_set}")
while work_set:
    item = work_set.pop()         # витягуємо по одному
    print(f"  витягнули: {item}, залишилось: {work_set}")
print(f"Після циклу: {work_set}")

try:
    work_set.pop()                # порожня множина → KeyError
except KeyError as e:
    print(f"pop() на порожній → KeyError: {e}")
````

## ⚙️ Частина 5: Алгоритми з циклом `while`

Хоча Python частіше використовує `for`, вивчення `while` з лічильником допомагає зрозуміти **механіку індексів** — основу всіх алгоритмів роботи з масивами.

### Шаблон

```python
i = 0                      # лічильник — починається з 0
while i < len(my_list):    # умова — не виходити за межі
    current = my_list[i]   # звертаємось до елемента за індексом
    # ... логіка ...
    i += 1                 # ОБОВ'ЯЗКОВО збільшуємо лічильник!
```

**Увага:** якщо забути `i += 1` — буде **нескінченний цикл**.

````python
# Алгоритм 5.1: Лінійний пошук елемента
import random

nums = [random.randint(1, 10) for _ in range(8)]
target = nums[random.randint(0, len(nums)-1)]   # беремо існуючий елемент
print(f"Список: {nums}")
print(f"Шукаємо: {target}")
print()

# Алгоритм
i = 0
found_indices = []

while i < len(nums):
    if nums[i] == target:
        found_indices.append(i)
    i += 1

if found_indices:
    print(f"Знайдено на індексах: {found_indices}")
    print(f"Перший індекс: {found_indices[0]}")
else:
    print("Не знайдено")

# Порівняння: вбудований метод
print(f"\nДля порівняння, nums.index({target}) = {nums.index(target)}  (тільки перший)")
print(f"nums.count({target}) = {nums.count(target)}")
````

````python
# Алгоритм 5.2: Копіювання з фільтрацією (безпечний спосіб)
# ВАЖЛИВО: ніколи не видаляйте елементи зі списку, по якому ітеруєтесь!
# Замість цього — збирайте результат у НОВИЙ список.

source = [15, 3, 8, 22, 7, 14, 1, 30, 11, 6]
print(f"Джерело: {source}")
print()

# Фільтр 1: парні числа
evens = []
i = 0
while i < len(source):
    if source[i] % 2 == 0:
        evens.append(source[i])
    i += 1
print(f"Парні числа: {evens}")

# Фільтр 2: числа > 10
big = []
i = 0
while i < len(source):
    if source[i] > 10:
        big.append(source[i])
    i += 1
print(f"Числа > 10:  {big}")

# Фільтр 3: перетворення (квадрати елементів < 10)
small_squares = []
i = 0
while i < len(source):
    if source[i] < 10:
        small_squares.append(source[i] ** 2)
    i += 1
print(f"Квадрати < 10: {small_squares}")
print()

print("=== Небезпечна помилка: видалення під час ітерації ===")
bad_list = [1, 2, 3, 4, 5, 6]
print(f"До видалення: {bad_list}")
# Небезпечно! Індекси зсуваються:
for item in bad_list[:]:
    if item % 2 == 0:
        bad_list.remove(item)
print(f"Після (copy): {bad_list}  ← правильний результат через копію зрізом")
````

````python
# Алгоритм 5.3: Ручне розвертання списку (Reverse in-place)
# Ідея: міняємо місцями елементи з протилежних кінців, рухаючись до центру

data = [1, 2, 3, 4, 5, 6, 7]
print(f"До розвертання: {data}")
print()

print("Покрокове виконання:")
left  = 0
right = len(data) - 1

while left < right:
    print(f"  Крок: data[{left}]={data[left]} <-> data[{right}]={data[right]}")
    data[left], data[right] = data[right], data[left]   # swap
    left  += 1
    right -= 1
    print(f"  Список: {data}")

print()
print(f"Після розвертання: {data}")
print()

# Порівняння зі вбудованими методами
original = [10, 20, 30, 40, 50]
print(f"original[::-1]   = {original[::-1]}  ← новий список")
original.reverse()                              # на місці
print(f"після .reverse() = {original}  ← на місці")
````

````python
# Алгоритм 5.4: Обчислення агрегатів (максимум, мінімум, середнє, підрахунок)
import random

scores = [random.randint(40, 100) for _ in range(10)]
print(f"Оцінки: {scores}")
print()

# Всі агрегати за ОДИН прохід
i         = 0
total_sum = 0
max_score = scores[0]    # початкове значення — перший елемент
min_score = scores[0]
above_70  = 0            # лічильник оцінок > 70

while i < len(scores):
    current = scores[i]

    total_sum += current

    if current > max_score:
        max_score = current

    if current < min_score:
        min_score = current

    if current > 70:
        above_70 += 1

    i += 1

average = total_sum / len(scores)

print(f"Максимум   : {max_score}")
print(f"Мінімум    : {min_score}")
print(f"Сума       : {total_sum}")
print(f"Середнє    : {average:.1f}")
print(f"Вище 70    : {above_70} студентів")
print()

print("=== Перевірка через вбудовані функції ===")
print(f"max(scores) = {max(scores)}, min(scores) = {min(scores)}, sum = {sum(scores)}")
````

````python
# Алгоритм 5.5: Злиття двох відсортованих списків (merge)
# Класичний алгоритм — два вказівники на два відсортованих списки

list_a = [1, 3, 5, 7, 9]
list_b = [2, 4, 6, 8, 10]
merged = []

i = 0   # вказівник на list_a
j = 0   # вказівник на list_b

while i < len(list_a) and j < len(list_b):
    if list_a[i] <= list_b[j]:
        merged.append(list_a[i])
        i += 1
    else:
        merged.append(list_b[j])
        j += 1

# Додаємо залишок
while i < len(list_a):
    merged.append(list_a[i])
    i += 1

while j < len(list_b):
    merged.append(list_b[j])
    j += 1

print(f"list_a  = {list_a}")
print(f"list_b  = {list_b}")
print(f"merged  = {merged}")
print(f"Правильно: {merged == sorted(list_a + list_b)}")
````

## ⚠️ Частина 6: Типові помилки початківців

| Помилка | Причина | Виправлення |
|---------|---------|-------------|
| `IndexError` | Індекс виходить за межі | Перевіряйте `i < len(list)` |
| `TypeError` (tuple) | Спроба змінити кортеж | Використайте список або створіть новий кортеж |
| `TypeError` (set) | Індексація множини | Конвертуйте `list(my_set)[0]` або ітеруйте |
| Зміна при ітерації | Індекси зсуваються | Ітеруйте по копії `list[:]` або збирайте в новий |
| `{}` замість `set()` | `{}` — це словник! | `set()` для порожньої множини |

````python
# Приклад 6.1: Демонстрація типових помилок

print("=== 1. IndexError ===")
lst = [10, 20, 30]   # індекси: 0, 1, 2
try:
    print(lst[3])    # помилка! максимальний індекс = 2
except IndexError as e:
    print(f"IndexError: {e}")
    print(f"  Список має {len(lst)} елементи, максимальний індекс = {len(lst)-1}")
print()

print("=== 2. TypeError — зміна кортежу ===")
tpl = (1, 2, 3)
try:
    tpl[0] = 99
except TypeError as e:
    print(f"TypeError: {e}")
    print("  Виправлення: list(tpl) → тепер можна змінювати")
    mutable = list(tpl)
    mutable[0] = 99
    print(f"  list(tpl) → {mutable}  ← ок")
print()

print("=== 3. TypeError — індексація множини ===")
my_set = {10, 20, 30}
try:
    _ = my_set[0]
except TypeError as e:
    print(f"TypeError: {e}")
    print("  Виправлення A: конвертуйте в список")
    print(f"  list(my_set)[0] = {list(my_set)[0]}  (але порядок довільний!)")
print()

print("=== 4. Зміна під час ітерації ===")
nums = [1, 2, 3, 4, 5, 6]
print(f"Оригінал: {nums}")

# Небезпечно (пропускає елементи!):
dangerous = [1, 2, 3, 4, 5, 6]
for n in dangerous:
    if n % 2 == 0:
        dangerous.remove(n)   # ← НЕБЕЗПЕЧНО
print(f"Небезпечно (пропуски): {dangerous}  ← 4 залишилась!")

# Безпечно (ітерація по копії):
safe = [1, 2, 3, 4, 5, 6]
for n in safe[:]:             # safe[:] — копія списку
    if n % 2 == 0:
        safe.remove(n)
print(f"Безпечно (копія):     {safe}  ← правильний результат")

# Ще краще — comprehension:
best = [n for n in [1,2,3,4,5,6] if n % 2 != 0]
print(f"Найкраще (comprehension): {best}")
print()

print("=== 5. {{}} — словник, не множина ===")
empty_dict = {}              # СЛОВНИК
empty_set  = set()           # МНОЖИНА
print(f"type({{}})      = {type(empty_dict).__name__}  ← СЛОВНИК!")
print(f"type(set()) = {type(empty_set).__name__}   ← множина (правильно)")
````

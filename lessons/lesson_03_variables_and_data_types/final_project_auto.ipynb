# 📦 Фінальний проєкт — Таємнича коробка Нової Пошти

Кожна група вирішує своє завдання і **лідер групи** відправляє результат.

```
         📦 BOX
           ↑
  ┌────────┼────────┬────────┐
  Група 1  Група 2  Група 3  Група 4
  Адреси   Вага    Логістика Відправка
```

> ⚠️ Якщо хоча б одна група помилиться — коробка не відкриється.

---
## 🚀 Фінальна перевірка

> ⚠️ **Цю клітинку запускає тільки викладач** після того, як всі групи відправили результати.

````python
# @title 🔒 Натисніть RUN ПЕРЕВІРКИ ШТРИХ КОДУ
# 🚀 ФІНАЛЬНА ПЕРЕВІРКА — тільки викладач

import requests
import csv
import io
import time
from IPython.display import display, HTML


# ===============================
# 📡 Google Sheet
# ===============================

FINAL_SHEET_CSV = (
    "https://docs.google.com/spreadsheets/d/e/2PACX-1vQ16oP8S-HwlXJsWhXtXpeUdIdZHV1mBnn9ipYQe0Dz57W4BnvGxkBG11NRatXY49SVOiGKFnFYe0Aw/pub?gid=2126041300&single=true&output=csv"
)

print("📡 Підключення до системи Нової Пошти...")
time.sleep(2)


# ===============================
# 📥 Завантаження результатів
# ===============================

results = {}

try:
    response = requests.get(FINAL_SHEET_CSV, timeout=10)

    reader = csv.DictReader(io.StringIO(response.text))

    for row in reader:

        # 🔥 нормалізація назв колонок
        clean_row = {
            k.strip().lower(): (v.strip() if v else "")
            for k, v in row.items()
        }

        group = clean_row.get("group")
        result = clean_row.get("result")

        if group in ["GROUP1", "GROUP2", "GROUP3", "GROUP4"] and result:
            results[group] = result

except Exception as e:
    print("❌ Помилка читання таблиці:", e)


print("\n📡 Отримано дані:")
print(results)

print("\n" + "=" * 45)
print("📦 ПЕРЕВІРКА КОРОБКИ")
print("=" * 45)

all_ok = True


# ===============================
# 👥 GROUP 1
# ===============================

print("\n🔍 Перевірка Адресного відділу...")
time.sleep(1)

g1 = results.get("GROUP1")

if str(g1) == "100":
    print("✅ Група 1 — Адреса OK")
else:
    print(f"❌ Помилка GROUP1 → {g1}")
    all_ok = False


# ===============================
# 👥 GROUP 2
# ===============================

print("\n🔍 Перевірка Відділу ваги...")
time.sleep(1)

try:
    if int(results.get("GROUP2")) == 25:
        print("✅ Група 2 — Вага OK")
    else:
        print("❌ GROUP2 — неправильна вага")
        all_ok = False
except:
    print("❌ GROUP2 — помилка типу")
    all_ok = False


# ===============================
# 👥 GROUP 3
# ===============================

print("\n🔍 Перевірка Логістики...")
time.sleep(1)

try:
    if int(results.get("GROUP3")) == 55:
        print("✅ Група 3 — Логістика OK")
    else:
        print("❌ GROUP3 — неправильна ціна")
        all_ok = False
except:
    print("❌ GROUP3 — помилка даних")
    all_ok = False


# ===============================
# 👥 GROUP 4
# ===============================

print("\n🔍 Перевірка Служби відправки...")
time.sleep(2)

EXPECTED_CITY_CODE = "[67, 104, 97, 114, 107, 105, 118]"
g4 = results.get("GROUP4")

if g4 == EXPECTED_CITY_CODE:
    print("✅ Група 4 — Код міста OK")
else:
    print("❌ GROUP4 — невірний код")
    print("Отримано:", g4)
    all_ok = False


# ===============================
# 🔓 FINAL BOX CHECK
# ===============================

print("\n" + "=" * 45)
time.sleep(2)

if all_ok:

    print("\n🚚 Передача підтверджена...")
    time.sleep(1)

    print("🔐 Перевірка завершена...")
    time.sleep(1)

    print("\n📡 Розшифровка місця відправлення...")
    time.sleep(2)

    received = [67, 104, 97, 114, 107, 105, 118]
    decoded = "".join(chr(x) for x in received)

    print("📍 Посилка прибула з:", decoded)
    time.sleep(1)
    print("\n🔓 КОРОБКУ ВІДКРИТО!")
    time.sleep(1)
    print("🎉 Усі відділи спрацювали правильно!")

    time.sleep(2)

    display(HTML("""
    <h1 style="color:#2ecc71;">🔓 BOX OPENED</h1>

    <a href="https://drive.google.com/drive/folders/101sjb2M_SL__TUlXeOK-K3Cjmkz64J3z?usp=sharing"
    target="_blank"
    style="
        display:inline-block;
        padding:20px 30px;
        font-size:24px;
        color:white;
        background:#27ae60;
        border-radius:12px;
        text-decoration:none;">
        📚 ACCESS MATERIALS
    </a>
    """))

else:
    print("🔒 Коробка залишається закритою.")
    print("   Знайдіть помилку та спробуйте ще раз.")

print("=" * 45)
````

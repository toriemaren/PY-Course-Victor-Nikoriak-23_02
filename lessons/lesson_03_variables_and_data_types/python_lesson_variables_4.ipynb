# Конспект: Змінні та примітивні типи даних у Python

---

## 1. Змінні (Variables)

Змінні у Python діють як **посилання** або *"ярлики"*, які вказують на об'єкти в пам’яті, а не як "коробки" для зберігання значень.

---

### 🔹 Створення змінної

Змінна створюється в момент присвоювання значення за допомогою оператора `=`.

```python
a = 5
````

У цьому випадку:

* створюється об'єкт цілого числа `5`
* ім'я `a` починає посилатися на цей об'єкт

---

### 🔹 Динамічна типізація

У Python **не потрібно оголошувати тип змінної наперед**.

Одна змінна може посилатися на різні типи даних у різний час:

```python
x = 10
x = "hello"
```

---

### 🔹 Правила іменування змінних

Імена можуть містити:

* літери
* цифри
* символ підкреслення `_`

Обмеження:

* ім'я **не може починатися з цифри**
* не допускаються пробіли
* Python чутливий до регістру

```python
spam != Spam
```

✅ Згідно зі стандартом **PEP 8**, використовують нижній регістр:

```python
user_age
total_sum
```

---

### 🔹 Копіювання змінних

```python
b = a
```

Це **не копіює об'єкт**, а створює нове посилання на той самий об'єкт у пам’яті.

---

## 2. Примітивні типи даних (Primitive Data Types)

Кожен об'єкт у Python має тип, який визначає можливі операції над ним.

---

### 🔹 `int` — цілі числа

```python
42
-5
```

У Python 3 мають практично необмежену довжину.

---

### 🔹 `float` — числа з плаваючою крапкою

```python
3.14
-1.2
```

---

### 🔹 `str` — рядки

Текстові дані (послідовність символів).

```python
"text"
'hello'
```

⚠️ **Рядки є незмінними (immutable).**

Після створення рядка його внутрішній вміст змінити неможливо.

#### Основні принципи незмінності рядків

**1. Заборона модифікації "на місці"**

```python
s = "abc"
s[0] = "x"
```

Викличе помилку:

```
TypeError: 'str' object does not support item assignment
```

---

**2. Створення нового об'єкта**

Будь-яка операція, що виглядає як зміна рядка, створює новий об'єкт:

```python
s = "hello"
s = s + " world"
```

Старий рядок не змінюється.

---

**3. Гарантія цілісності даних**

Незмінність дозволяє використовувати рядки як:

* ключі словників (`dict`)
* елементи множин (`set`)

Їхній хеш залишається стабільним.

---

**4. Оптимізація пам'яті**

Рядки зберігаються компактно та ефективно завдяки незмінності.

---

✅ Найближчий змінний аналог рядка:

* `list`
* `bytearray`

---

### 🔹 `bool` — логічний тип

```python
True
False
```

---

### 🔹 `None`

Спеціальне значення, що означає **відсутність значення**.

```python
value = None
```

---

## 3. Оператори та вирази (Operators)

---

### 🔹 Математичні оператори

| Оператор | Опис                  |
| -------- | --------------------- |
| `+`      | додавання             |
| `-`      | віднімання            |
| `*`      | множення              |
| `**`     | піднесення до степеня |

---

### 🔹 Ділення

#### `/` — стандартне ділення

Завжди повертає `float`.

```python
10 / 4  # 2.5
```

---

#### `//` — цілочисельне ділення

Відкидає дробову частину.

```python
10 // 4  # 2
```

---

#### `%` — остача від ділення

```python
10 % 4  # 2
```

---

### 🔹 Операції з рядками

#### Конкатенація

```python
"Hello" + " World"
```

#### Повторення

```python
"Hi" * 3
```

---


### 4🔹 Чому Python вимагає явного перетворення типів?

Python поєднує в собі **динамічну**, але **строгу** типізацію.

Це означає:

- тип змінної **не потрібно оголошувати заздалегідь** (динамічність)
- але кожен об'єкт у пам’яті **завжди має чітко визначений тип**
- Python суворо контролює, які операції дозволені для цього типу

---

#### 🧠 Філософія Python

Один із головних принципів Python:

> **Explicit is better than implicit**  
> (*Явне краще за неявне*)

На відміну від мов типу JavaScript або PHP, Python **майже ніколи не виконує автоматичне перетворення типів**.

---

### Приклад

У JavaScript:

```javascript
"spam" + 5
````

✅ результат:

```
"spam5"
```

У Python:

```python
"spam" + 5
```

❌ результат:

```
TypeError: can only concatenate str (not "int") to str
```

Python не намагається "вгадати", що саме ви хотіли зробити.

---

### ✅ Чому це добре?

#### 1. Уникнення неоднозначності

Якщо є:

```python
"42" + 5
```

Python не знає:

* чи потрібно отримати число `47`
* чи текст `"425"`

Тому він зупиняє виконання.

---

#### 2. Запобігання прихованим помилкам

Явне перетворення змушує програміста чітко визначати намір:

```python
int("42") + 5
str(42) + "5"
```

Помилка `TypeError` одразу показує логічну проблему в коді.

---

## 🦆 Duck Typing (Утина типізація)

Щоб компенсувати строгість типів, Python використовує концепцію
**утиної типізації (Duck Typing)**.

Вона базується на відомому принципі:

> *"Якщо воно виглядає як качка, плаває як качка і крякає як качка — це качка."*

---

### Що це означає в Python?

Python **не перевіряє походження об'єкта**,
його цікавить лише **поведінка**.

Не важливо:

* до якого класу належить об'єкт
* хто його батьківський клас

Важливо лише:

✅ чи підтримує він потрібні операції.

---

### Приклад: ітерація

Функції байдуже, що саме передати:

```python
for item in obj:
    print(item)
```

Це може бути:

* `list`
* `tuple`
* `str`
* генератор
* власний клас

Головне — щоб об'єкт підтримував ітерацію.

---

### Приклад: оператор `+`

Об'єкту не обов'язково бути числом.

Достатньо реалізувати магічний метод:

```python
__add__()
```

---

## ⚖️ Баланс Python

Python одночасно:

✅ **строгий** — не дозволяє змішувати несумісні типи

✅ **гнучкий** — дозволяє працювати з будь-якими об'єктами, що поводяться правильно

---

Python контролює *що* ви робите з даними,
але майже не обмежує *якими саме об'єктами* ви це робите.

---

### 🔹 `int(x)`

```python
int("10")
int(3.9)  # 3
```

---

### 🔹 `float(x)`

```python
float(5)
float("3.14")
```

---

### 🔹 `str(x)`

```python
str(100)
```

---

## Висновок

У Python:

* змінні — це посилання на об'єкти
* тип визначається автоматично
* правильна робота з даними базується на розумінні типів і операцій над ними

# 🔤 Робота з символами в Python: індекси, `ord()` та `chr()`

Цей урок показує **як Python працює з текстом на рівні символів**.  
Це фундамент для:

- ✅ рядків  
- ✅ кодування
- ✅ шифрування
- ✅ NLP
- ✅ AI алгоритмів

---
## 🐍 1. Рядок — це послідовність символів

Python зберігає рядок як пронумеровану послідовність:

```
P   y   t   h   o   n
0   1   2   3   4   5
```

Доступ до символів — через **індекс** у квадратних дужках.

`````python
word = 'Python'

print("Рядок:", word)
print("word[0]:", word[0])   # перший символ
print("word[1]:", word[1])   # другий символ
print("word[5]:", word[5])   # останній символ (індекс 5)
`````

---
## 🔙 Негативні індекси

Python дозволяє рахувати **з кінця**:

```
P    y    t    h    o    n
-6  -5   -4   -3   -2   -1
```

Зручно для: останнього символу, розширення файлу, кінця рядка.

`````python
word = 'Python'

print("word[-1]:", word[-1])   # останній
print("word[-2]:", word[-2])   # передостанній
print("word[-6]:", word[-6])   # те саме, що word[0]
`````

---
## 🔢 2. Символ = число

Комп'ютер не розуміє літери — він працює лише з числами.  
Кожен символ має числовий код (**Unicode / ASCII**).

| Функція | Що робить           | Приклад              |
|---------|---------------------|----------------------|
| `ord()` | символ → число      | `ord('A')` → `65`    |
| `chr()` | число → символ      | `chr(66)` → `'B'`    |

Це **зворотні операції**:
```python
chr(ord('A')) == 'A'   # True
```

`````python
# ord() — символ → число
print("ord('A'):", ord('A'))
print("ord('a'):", ord('a'))
print("ord('z'):", ord('z'))

print()

# chr() — число → символ
print("chr(65):", chr(65))
print("chr(66):", chr(66))
print("chr(90):", chr(90))

print()

# Зворотність
print("chr(ord('A')) == 'A':", chr(ord('A')) == 'A')
`````

---
## 🔐 Для чого це використовується

### 1. Шифрування (Caesar cipher) — зсунути букву
```python
chr(ord('A') + 1)   # → 'B'
```

### 2. Перевірка типу символу
```python
if ord(char) >= ord('a') and ord(char) <= ord('z'):
    ...  # маленька літера
```

### 3. Генерація алфавіту
```python
for i in range(ord('a'), ord('z') + 1):
    print(chr(i))
```

`````python
# 1. Зсув літери (шифрування)
print("A + 1 →", chr(ord('A') + 1))
print("a + 3 →", chr(ord('a') + 3))

print()

# 2. Перевірка: чи символ — маленька літера?
char = 'g'
if ord(char) >= ord('a') and ord(char) <= ord('z'):
    print(f"'{char}' — маленька літера ✅")

print()

# 3. Генерація першых 5 літер алфавіту
print("Перші 5 літер:", end=" ")
for i in range(ord('a'), ord('a') + 5):
    print(chr(i), end=" ")
print()
`````
